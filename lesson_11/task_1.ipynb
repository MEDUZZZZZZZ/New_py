Дана функция: f(x) = sin(x)^2 - cos(x)^2  

Для нее:
> 1. Определить корни
> 2. Найти интервалы, на которых функция возрастает
> 3. Найти интервалы, на которых функция убывает
> 4. Построить график
> 5. Вычислить вершину
> 6. Определить промежутки, на котором f > 0
> 7. пределить промежутки, на котором f < 0

```python
import sympy as sm
from sympy import S
x = sm.Symbol('x')
my_func = sm.sin(x)**2 - sm.cos(x)**2
my_func
```

**Определение корней**

```python
roots = sm.solveset(my_func, x)
roots
```

**График функции**

```python
pic = sm.plot(my_func, line_color ="purple")
```

**Промежутки где функция больше 0**

```python
more_null = sm.solveset(my_func > 0, x, domain = S.Reals)
more_null
```

**Промежутки где функция меньше нуля 0**


```python
less_null = sm.solveset(my_func < 0, x, domain = S.Reals)
less_null
```

**Вершины функции**

```python
funcc_diff = sm.diff(my_func, x)
extr = sm.solveset(funcc_diff, x, domain = S.Reals)
extr
```

**Интервал возрастания**

```python
grow = sm.solveset(funcc_diff > 0, x, domain = S.Reals)
grow
```

**Интервал убывания**

```python
fall = sm.solveset(funcc_diff < 0, x, domain = S.Reals)
fall
```
